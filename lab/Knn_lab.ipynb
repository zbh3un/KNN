```python
from keras.datasets import mnist
df = mnist.load_data('minst.db')
train,test = df
X_train, y_train = train
X_test, y_test = test
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000)
for i in range(5):
    print(y_test[i],'\n') # Print the label
    print(X_test[i],'\n') # Print the matrix of values
    plt.contourf(np.rot90(X_test[i].transpose())) # Make a contour plot of the matrix values
    plt.show()
```

```python
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("X_train[0] shape:", X_train[0].shape)
print("X_test[0] shape:", X_test[0].shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
X_train_flat = X_train.reshape(X_train.shape[0], 784)
X_test_flat = X_test.reshape(X_test.shape[0], 784)

print("X_train_flat shape:", X_train_flat.shape)
print("X_test_flat shape:", X_test_flat.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

# Use a subset for faster cross-validation
sample_size = 5000
np.random.seed(42)
idx = np.random.choice(X_train_flat.shape[0], sample_size, replace=False)
X_sample = X_train_flat[idx]
y_sample = y_train[idx]

k_values = [1, 3, 5, 7, 9]
cv_scores = {}

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_sample, y_sample, cv=5, scoring='accuracy')
    cv_scores[k] = scores.mean()
    print(f"k={k}: CV Accuracy = {scores.mean():.4f}")

optimal_k = max(cv_scores, key=cv_scores.get)
print(f"\nOptimal k = {optimal_k}")
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Train on FULL training set with optimal k
knn_best = KNeighborsClassifier(n_neighbors=optimal_k)
knn_best.fit(X_train_flat, y_train)

y_pred = knn_best.predict(X_test_flat)

print("Test Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
import seaborn as sns

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=range(10), yticklabels=range(10))
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
